```python
from render import Render
from pyflux import Flux
```

```python
host = 'hackmilan.a.influxcloud.net'
version = 'v2'
```

```python
f = Flux(host, 8080, version)
render = Render()
```

```python
results = f.eval_query('from(bucket: "telegraf") |> range(start:-3m) |> filter(fn: (r) => r._measurement == "cpu" and r._field == "usage_system" and r.cpu=="cpu0")')
```

```python
%matplotlib inline

render.timeplot(results)
```
